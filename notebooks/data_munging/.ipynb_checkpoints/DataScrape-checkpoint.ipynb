```python
from os import listdir
from os.path import isfile,join
import pandas as pd
```

```python
paths=['../../data/y2018/','../../data/y2019/']
years=[2018,2019]
```

```python
i=0
for path,year in zip(paths,years):
    print(path)
    print(year)
    file_list=listdir(path)
    
    for file in file_list:
        if i==0:
            i+=1
            all_files=pd.read_csv(path+file)
            all_files['YEAR']=year
        else:
            cur_file=pd.read_csv(path+file)
            cur_file['YEAR']=year
            
            all_files=pd.concat([all_files,cur_file])
```

```python
print(all_files.shape)
all_files.head()
```

```python
all_files.groupby(['OFF_TEAM_ABBREVIATION','DEF_TEAM_ABBREVIATION','YEAR']).count().sort_values('OFF_PLAYER_ID')
```

```python
all_files.to_csv('../../data/all_matchups.csv')
```
